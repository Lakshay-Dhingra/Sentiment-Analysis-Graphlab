```python
selected_words = ['awesome', 'great', 'fantastic', 'amazing', 'love', 'horrible', 'bad', 'terrible', 'awful', 'wow', 'hate']
```

```python
for i in selected_words:
    def my_count(word_count):
        if i in word_count:
            return word_count[i]
        return 0
    products[i]=products['word_count'].apply(my_count)
```

### 2. Create a new sentiment analysis model using only the selected_words as features: In the Jupyter Notebook above, we used word counts for all words as features for our sentiment classifier. Now, you are just going to use the selected_words.

```python
#Ignore 3 star reviews
products=products[products['rating']!=3.0]
#positive sentiment = 4 or 5 star review
#negative sentiment = 1 or 2 star review
products['sentiment']=products['rating']>=4
```

```python
train_data,test_data = products.random_split(.8, seed=0)
selected_words_model=graphlab.logistic_classifier.create(train_data,target='sentiment',features=selected_words,validation_set=test_data)
```

```python
# You will now examine the weights the learned classifier assigned to each of the 11 words in selected_words and 
# gain intuition as to what the ML algorithm did for your data using these features. In Turi Create, a learned 
# model, such as the selected_words_model, has a field 'coefficients', which lets you look at the learned 
# coefficients.

selected_words_model['coefficients']

# The result has a column called ‘value’, which contains the weight learned for each feature.
```

```python
selected_words_model['coefficients'].sort('value').head()
```

### 3. Comparing the accuracy of different sentiment analysis model

```python
selected_words_model.evaluate(test_data)
```

```python
# Accuracy in case of majority class predictor
# 1 or positive review is majority class
pos=27846+130
neg=234+5094
majority_accuracy=float(pos)/float(pos+neg)
print majority_accuracy
```
